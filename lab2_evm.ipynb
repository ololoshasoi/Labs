```python
def count_K (matr, ver_num, ro):
    K = []
    for i in range (len(matr)):
        temp_K = 0
        if i not in ver_num:
            K.append(None)
            continue
            
        temp_K -= ro[i] #Отнимаем сумму всех связей
        for j in range (len(matr)):
            if j not in ver_num:
                temp_K += 2*matr[i][j]
        K.append(temp_K)
    return(K)

def max_K (K):
    max_k = -100000000000000000000
    for i in range(len(K)):
        if (K[i]!= None):
            if (K[i] > max_k):
                max_k = K[i]
                pos = i
    return (pos, max_k)
            
def count_Q (matr, pos):
    Q = 0
    for i in range(len(matr)):
        for j in range (i, len(matr)):
            if (matr[i][j] > 0):
                list_index_first =[ [k,f] for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == i]
                list_index_second =[ [k,f] for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == j]
                dist = abs(list_index_first[0][0] - list_index_second[0][0]) + abs(list_index_first[0][1] - list_index_second[0][1])
                Q += dist*matr[i][j]
    
    return(Q)

def count_L (matr, S_V, pos):
    L = []
    for i in range(len(matr)):
        temp_L = 1
        summa = 0
        temp_L = temp_L / S_V[i]
        for j in range(len(matr[i])):
             if (matr[i][j] > 0):
                list_index_first =[ [k,f] for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == i]
                list_index_second =[ [k,f] for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == j]
                dist = abs(list_index_first[0][0] - list_index_second[0][0]) + abs(list_index_first[0][1] - list_index_second[0][1])
                #print(matr[i][j])
                #print(dist)
                summa += dist*matr[i][j]           
        temp_L = temp_L * summa
        #print("kek")
        
        L.append(temp_L)
    return (L)
                       

def count_delt_L(x, y, max_L_pos, L, matr, positions, S_V, N):
    new_x = int(x - 1)
    new_y = int(N - y)
    num_elem = int(positions[new_y][new_x])
    delt_L = 0 - L[max_L_pos] - L[num_elem]
    pos = copy.deepcopy(positions)
    list_index =[ [k,f] for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == max_L_pos]
    elem_x = list_index[0][1]
    elem_y = list_index[0][0]
    pos[elem_y][elem_x], pos[new_y][new_x] = pos[new_y][new_x], pos[elem_y][elem_x]
    new_L = count_L(matr, S_V, pos)
    delt_L = delt_L + new_L[num_elem] + new_L[max_L_pos]
    return (delt_L)
    
def count_x (matr, S_V, pos, num):
    temp_x = 1
    summa = 0
    temp_x = temp_x / S_V
    for j in range(len(matr[num])):
        if (matr[num][j] > 0):
            list_index_first =[ f for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == j]
            summa += (list_index_first[0]+1) * matr[num][j]  
    temp_x = temp_x * summa
    return (temp_x)

def count_y (matr, S_V, pos, num, N):
    temp_y = 1
    summa = 0
    temp_y = temp_y / S_V
    for j in range(len(matr[num])):
        if (matr[num][j] > 0):
            list_index_first =[ k for k in range(len(pos))  for f in range(len(pos[k])) if pos[k][f]  == j]
            summa += (N - list_index_first[0]) * matr[num][j]    
    temp_y = temp_y * summa
    return (temp_y)
```
