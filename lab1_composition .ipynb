```python
def summ(matr, hor, start_c, end_c):
    summa = 0
    for i in range(start_c, end_c):
        summa += matr[hor][i]
    return(summa)

def count_Q (matr, variants):
    displacement = 0 #смещение
    Q = 0
    for container in variants:
        for i in range (displacement, displacement+container):  
            for j in range (displacement+container, len(matr)):
                Q = Q + matr[i][j]
        displacement = displacement + container
    return (Q)

def count_delt_R (matr, container , cont_num, variants):
    delt_R = []
    colum = []
    colum.append(0) #старт контейнера
    displacement = 0
    for i in range (len(variants)):
        colum.append(colum[i]+variants[i]) 
    for i in range (cont_num):  
        displacement += variants[i]
    #Расчет дельта R
    start_hor = displacement - container 
    z = 0 
    for i in range (start_hor, displacement):
        delt_R.append([])
        for j in range (displacement, len(matr)): 
            for k in range (len(colum)):
                if j >= colum[k]:
                    start_cont = colum[k]
                    end_cont = colum[k+1]
                    
            R = summ(matr, i, start_cont, end_cont) - summ(matr, i, start_hor, displacement) + summ(matr, j, start_hor, displacement) - summ(matr, j, start_cont, end_cont) - 2*matr[i][j]
            delt_R[z].append(R)
        z = z+1
    return(delt_R)

def find_max_R(delt_R):
    max_R = delt_R[0][0]
    max_hor = 0
    max_ver = 0
    for i in range (0, len(delt_R)):
        for j in range (0,len(delt_R[0])):
            if (delt_R[i][j] > max_R):
                max_R = delt_R[i][j]
                max_hor = i
                max_ver = j
    return (max_R, max_hor, max_ver)

def find_min_S (S_V, matr):
    min_S = S_V[0] 
    min_num_S = 0 
        
    for i in range (1, len(S_V)):
        if (S_V[i] < min_S): 
            min_S = S_V[i]
            min_num_S = i
        if (S_V[i] == min_S): 
            s_min = 0
            s_i = 0
            for j in range (len(matr[i])):
                if (matr[i][j] > 1):
                    s_i = s_i + 1
                if (matr[min_num_S][j] > 1):
                    s_min = s_min + 1
            if (s_i > s_min):
                min_S = S_V[i]
                min_num_S = i 
                
    return (min_num_S)
```
